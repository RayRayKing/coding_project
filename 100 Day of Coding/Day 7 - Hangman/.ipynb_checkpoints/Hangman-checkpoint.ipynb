```python
import random

stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']


word_list = ["ardvark","baboon","camel"]

word_choice = random.choice(word_list)
print(f"Your word choice is {word_choice}")

display = list()

for space in range(len(word_choice)):
    display.append('_')
    
lives = 6
end_of_game = False
while not end_of_game:    
    guess = input("What is your letter?\n").lower()
    if guess in word_choice:
        for idx,ltr in enumerate(word_choice):
            if guess == ltr:
                display[idx] = ltr
        if "_" not in display:
            end_of_game = True
            print("You Win!")
    else:
        lives -= 1
    if lives == 0:
        end_of_game = True
        print("You Lose!")

    print(display)
    print(stages[lives])


        
# check input against variable
```
